## Logistic Regression (From scratch)

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))


def compute_cost_logistic(X, y, w, b, lam):

    m = X.shape[0]  # number of training examples
    z = np.dot(X, w) + b  # (m,1)
    h = sigmoid(z)  # (m,1)

    cost = -(1.0/m) * np.sum(y * np.log(h) + (1.0 - y) * np.log(1.0 - h))
    
    # L2 regularization term to the cost (excluding bias 'b')
    cost += (lam / (2.0 * m)) * np.sum(w**2) 
    
    return cost


def compute_gradient_logistic(X, y, w, b, lam):

    m = X.shape[0]
    z = np.dot(X, w) + b  # (m,1)
    h = sigmoid(z)  # (m,1)
    err = h - y  # (m,1)

    dj_dw = (1.0/m) * np.dot(X.T, err) + (lam / m) * w  # L2 regularization term added to dj_dw
    dj_db = (1.0/m) * np.sum(err)  # scalar
    return dj_dw, dj_db


def gradient_descent_logistic(X, y, w_in, b_in, alpha, num_iters, lam):
    w = w_in.copy()  
    b = float(b_in) 
    J_history = []  # To store cost values during training

    for i in range(num_iters):
        # Compute gradients
        dj_dw, dj_db = compute_gradient_logistic(X, y, w, b, lam)
        
        # Update weights and bias
        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        # Record cost every 100 iterations (or at the final iteration)
        if (i % 100 == 0) or (i == num_iters - 1):
            J_history.append(compute_cost_logistic(X, y, w, b, lam))
            print(f"Iteration {i}: Cost={J_history[-1]:.6f}, b={b:.4f}, w={w.T}")

    return w, b, J_history



def predict_logistic(X, w, b, threshold=0.5):
    proba = sigmoid(np.dot(X, w) + b)  # Probability predictions
    return (proba >= threshold).astype(int)  # Convert to binary labels (0 or 1)
```
